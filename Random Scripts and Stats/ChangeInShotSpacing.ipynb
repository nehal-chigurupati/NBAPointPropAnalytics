```python
import csv
```

```python
fileIn = open('ClosestDefenderStats.csv')
csvreader = csv.reader(fileIn)
```

```python
Results = {}
```

```python
Results["header"] = next(csvreader)
```

```python
print(Results)
```

```python
Results["22-23"] = next(csvreader)
Results["21-22"] = next(csvreader)
Results["20-21"] = next(csvreader)
Results["19-20"] = next(csvreader)
Results["18-19"] = next(csvreader)
Results["17-18"] = next(csvreader)
```

```python
X = ["17-18", "18-19", "19-20", "20-21", "21-22", "22-23"]
```

```python
veryTightFreqs = [float(Results["17-18"][2]) * 100, float(Results["18-19"][2]) * 100,float(Results["19-20"][2]) * 100, float(Results["20-21"][2]) * 100, float(Results["21-22"][2]) * 100, float(Results["22-23"][2]) * 100]
```

```python
print(veryTightFreqs)
```

```python
tightFreqs = [float(Results["17-18"][7]) * 100, float(Results["18-19"][7]) * 100,float(Results["19-20"][7]) * 100, float(Results["20-21"][7]) * 100, float(Results["21-22"][7]) * 100, float(Results["22-23"][7]) * 100]
```

```python
openFreqs = [float(Results["17-18"][12]) * 100, float(Results["18-19"][12]) * 100,float(Results["19-20"][12]) * 100, float(Results["20-21"][12]) * 100, float(Results["21-22"][12]) * 100, float(Results["22-23"][12]) * 100]
```

```python
wideOpenFreqs = [float(Results["17-18"][17]) * 100, float(Results["18-19"][17]) * 100,float(Results["19-20"][17]) * 100, float(Results["20-21"][17]) * 100, float(Results["21-22"][17]) * 100, float(Results["22-23"][17]) * 100]
```

```python
print(wideOpenFreqs)
```

```python
freqs = {"0-2 feet": veryTightFreqs, "2-4 feet": tightFreqs, "4-6 feet": openFreqs, "6+ feet": wideOpenFreqs}
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()
bottom = np.zeros(6)
for boolean, count in freqs.items():
    p = ax.bar(X, count, .5, label=boolean, bottom=bottom)
    bottom += count
ax.set_title("James Harden Shot Spacing Distributions")
ax.legend(title = "Distance to Nearest Defender", bbox_to_anchor=(1.1, 1.05))
plt.xlabel("Season")
plt.ylabel("Percent of Shot Attempts")
plt.show()
```
